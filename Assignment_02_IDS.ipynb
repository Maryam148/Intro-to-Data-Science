```python
import random

# PART 1

users = [
    (1, "Alice", 25, "USA"),
    (2, "Bob", 35, "Canada"),
    (3, "Charlie", 28, "UK"),
    (4, "David", 32, "Australia"),
    (5, "Emily", 27, "Germany"),
    (6, "Frank", 40, "France"),
    (7, "Grace", 23, "Spain"),
    (8, "Henry", 38, "Italy"),
    (9, "Ivy", 31, "China"),
    (10, "Jack", 26, "Japan"),
    (11, "Katherine", 34, "Brazil"),
    (12, "Liam", 29, "Russia"),
    (13, "Maria", 37, "India"),
    (14, "Noah", 22, "Mexico"),
    (15, "Olivia", 30, "South Africa"),
    (16, "Paul", 36, "Nigeria"),
    (17, "Quinn", 28, "Egypt"),
    (18, "Riley", 33, "Argentina"),
    (19, "Sarah", 27, "Colombia"),
    (20, "Thomas", 35, "Canada"),
    (21, "Uma", 32, "Turkey"),
    (22, "Victor", 29, "Indonesia"),
    (23, "Wendy", 38, "Malaysia"),
    (24, "Xavier", 26, "Philippines"),
    (25, "Yara", 34, "Vietnam"),
    (26, "Zachary", 31, "Thailand"),
    (27, "Abigail", 28, "Saudi Arabia"),
    (28, "Benjamin", 37, "Iran"),
    (29, "Chloe", 25, "Iraq"),
    (30, "Daniel", 33, "Israel"),
    (31, "Eleanor", 30, "Lebanon"),
    (32, "Finn", 36, "USA"),
    (33, "Gabriela", 29, "Qatar"),
    (34, "Harold", 38, "Syria"),
    (35, "Isabella", 27, "Tunisia"),
    (36, "Jacob", 35, "Canada"),
    (37, "Katherine", 32, "Algeria"),
    (38, "Liam", 29, "Angola"),
    (39, "Maria", 38, "Botswana"),
    (40, "Noah", 26, "Cameroon")]

# def filter_users(users, countries, age_limit):
#     filtered_users = []
#     for user in users:
#         if user[3] in countries and user[2] > age_limit:
#             filtered_users.append(user)
#     return [user[1] for user in filtered_users]

# filtered_names = filter_users(users, ["USA", "Canada"], 30)
# print(filtered_names)

def sort_and_duplicates(users):
    sorted_users = sorted(users, key=lambda x: x[2], reverse=True)
    top_10_oldest = sorted_users[:10]
    duplicate_names = set()
    name_set = set()
    for user in users:
        name = user[1]
        if name in name_set:
            duplicate_names.add(name)
        else:
            name_set.add(name)
    return top_10_oldest, duplicate_names


top_10, duplicates = sort_and_duplicates(users)
print(duplicates)

# PART 2

transactions = [
    (1, 123, 100.0, "2023-01-01"),
    (2, 456, 50.0, "2023-02-02"),
    (3, 123, 20.0, "2023-03-03"),
    (4, 789, 30.0, "2023-04-04"),
    (5, 123, 40.0, "2023-05-05"),
    (6, 987, 60.0, "2023-06-06"),
    (7, 123, 70.0, "2023-07-07"),
    (8, 321, 80.0, "2023-08-08"),
    (9, 123, 90.0, "2023-09-09"),
    (10, 654, 100.0, "2023-10-10"),
    (11, 123, 110.0, "2023-11-11"),
    (12, 543, 120.0, "2023-12-12"),
    (13, 123, 130.0, "2024-01-01"),
    (14, 432, 140.0, "2024-02-02"),
    (15, 123, 150.0, "2024-03-03"),
    (16, 321, 160.0, "2024-04-04"),
    (17, 123, 170.0, "2024-05-05"),
    (18, 213, 180.0, "2024-06-06"),
    (19, 123, 190.0, "2024-07-07"),
    (20, 654, 200.0, "2024-08-08")]

def count_unique_users(transactions):
    user_ids = set()
    for transaction in transactions:
        user_ids.add(transaction[1])
    return len(user_ids)

nmbrOfUniqueUsrs = count_unique_users(transactions)
print(nmbrOfUniqueUsrs)


def highest_amount_transaction(transactions):
    highest_amount = 0
    highest_transaction = None
    for transaction in transactions:
        if transaction[2] > highest_amount:
            highest_amount = transaction[2]
            highest_transaction = transaction
    return highest_transaction
 
highest_t=highest_amount_transaction(transactions)
print(highest_t)

def separate_transaction_and_user_ids(transactions):
    transaction_ids = []
    user_ids = []
    for transaction in transactions:
        transaction_ids.append(transaction[0])
        user_ids.append(transaction[1])
    return transaction_ids, user_ids

t_id, usr_id = separate_transaction_and_user_ids(transactions)
print(t_id,usr_id)

#PART 3

page_a_visitors = {
    'Alice', 'Bob', 'Charlie', 'Daisy', 'Eva', 
    'Frank', 'Grace', 'Henry', 'Isabella', 'Jack',
    'Katherine', 'Liam', 'Mia', 'Noah', 'Olivia'
}

page_b_visitors = {
    'Bob', 'David', 'Ethan', 'Henry', 'Isabella',
    'Jacob', 'Liam', 'Mason'
}

page_c_visitors = {
    'Alice', 'Sophia', 'Daisy', 'Grace', 'Oliver',
    'Paula', 'Quinn', 'Rita'
}

new_visits_a = {'Sophie', 'Tom', 'Uma'}
remove_visits_b = {'Bob', 'Ethan', 'Isabella', 'Mason'}

def manage_user_visits(page_a, page_b, page_c, new_visits_a, remove_visits_b):

    visited_ab = page_a.intersection(page_b)
    print("Users who visited both Page A and Page B:", visited_ab)
    visited_a_or_c = page_a.symmetric_difference(page_c) #symmetric_difference is used to find the elements that are in either of the two sets but not in both.
    print("Users who visited either Page A or Page C, but not both:", visited_a_or_c)

    page_a.update(new_visits_a) #update allows you to add multiple elements at once.
    print("Updated user names for Page A:", page_a)

    page_b.difference_update(remove_visits_b) #difference_update is used to remove the elements from set
    print("Updated user names for Page B after removal:", page_b)

manage_user_visits(page_a_visitors, page_b_visitors, page_c_visitors, new_visits_a, remove_visits_b)


# PART 4

feedback_data = {
    'Alice': {'rating': 5, 'comments': 'Excellent service!'},
    'Bob': {'rating': 4, 'comments': 'Very good, will return!'},
    'Charlie': {'rating': 3, 'comments': 'It was okay.'},
    'Daisy': {'rating': 5, 'comments': 'Loved it!'},
    'Eva': {'rating': 2, 'comments': 'Not great.'},
    'Frank': {'rating': 4, 'comments': 'Pretty good!'},
    'Grace': {'rating': 3, 'comments': 'Average experience.'},
}
feedback_data_2 = {
    'Alice': {'rating': 4, 'comments': 'Still good!'},
    'Charlie': {'rating': 5, 'comments': 'Improved a lot!'},
    'Hannah': {'rating': 4, 'comments': 'Great service!'},
}


def filter_and_sort_feedback(feedback):
     # Initialize an empty dictionary to store filtered feedback
    filtered_feedback = {}

    # Filter users who rated 4 or higher
    for user_id, details in feedback.items():
        if details['rating'] >= 4:
            filtered_feedback[user_id] = details['rating']
    
    # Sort the filtered dictionary by rating in descending order
    sorted_feedback = sorted(filtered_feedback.items(), key=lambda item: item[1], reverse=True)

    # Initialize an empty dictionary for the top 5 users
    top_5_users = {}

    # Iterate over the first 5 items in the sorted_feedback list
    for user_id, rating in sorted_feedback[:5]:
        top_5_users[user_id] = rating
    
    return top_5_users

def combine_feedback(feedback_list):
    combined_feedback = {}

    for feedback in feedback_list:
        for user_id, details in feedback.items():
            if user_id in combined_feedback:
                if details['rating'] > combined_feedback[user_id]['rating']:
                    combined_feedback[user_id]['rating'] = details['rating']
                combined_feedback[user_id]['comments'].append(details['comments'])
            else:
                combined_feedback[user_id] = {
                    'rating': details['rating'],
                    'comments': [details['comments']]
                }

    rating_above_3 = {}
    for user_id, details in combined_feedback.items():
        if details['rating'] > 3:
            rating_above_3[user_id] = details['rating']
    return combined_feedback,rating_above_3

top_users = filter_and_sort_feedback(feedback_data)
print("Top 5 Users with Ratings 4 or Higher:", top_users)

combined_feedback, rating_abv3= combine_feedback([feedback_data, feedback_data_2])
print("Combined Feedback:", combined_feedback)
print("User IDs with Ratings Greater Than 3:", rating_abv3)


```
